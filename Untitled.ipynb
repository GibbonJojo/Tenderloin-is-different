For this, Bar plots are my choice. Also I will look at the Category column more closely. But a plot for that would be pretty messy (too many unique values), so I will just display text

```python
def barplot(x):
    plt.bar(sf[x].value_counts().index, 
            sf[x].value_counts())
    plt.xticks(rotation="vertical")
    plt.show()
```

```python
for x in ["DayOfWeek", "PdDistrict"]:
    barplot(x)
display(sf["Category"].value_counts())
```

I also want to take a look at the amount of crime dependent on the time of day and month. For that, I will append two extra columns to the dataframe, containing only the hour and the month

```python
sf["hour"] = [datetime.strptime(x, "%H:%M").hour for x in sf["Time"]]
sf["month"] = [datetime.strptime(x, "%m/%d/%Y").month for x in sf["Date"]]
display(sf.head())
sf["month"].value_counts()
```

So, apparently the dataset only contains 3 months (June, July and August), which is unfortunate.

But we can still discover the timeseries of crimes:

```python
plt.bar(sf["hour"].value_counts().index, 
        sf["hour"].value_counts())
plt.xticks(sf["hour"].value_counts().index)
plt.show()
```

Here, we see that most crimes occur in the evening and early night. In the morning, crime rate drops rapidly and rises soon again.

We can confirm that, by grouping the times into periods:
Morning (5-10), Day (11-16), Evening (17-22) and Night (23-4)

```python
#plt.bar()
sf["Timeperiod"] = "None"
sf["Timeperiod"].loc[(sf["hour"] == 17) | (sf["hour"] == 18) | (sf["hour"] == 19) | (sf["hour"] == 20) | (sf["hour"] == 21) | (sf["hour"] == 22)] = "Evening (17-22)"
sf["Timeperiod"].loc[(sf["hour"] == 23) | (sf["hour"] == 0) | (sf["hour"] == 1) | (sf["hour"] == 2) | (sf["hour"] == 3) | (sf["hour"] == 4)] = "Night (23-4)"
sf["Timeperiod"].loc[(sf["hour"] == 5) | (sf["hour"] == 6) | (sf["hour"] == 7) | (sf["hour"] == 8) | (sf["hour"] == 9) | (sf["hour"] == 10)] = "Morning (5-10)"
sf["Timeperiod"].loc[(sf["hour"] == 11) | (sf["hour"] == 12) | (sf["hour"] == 13) | (sf["hour"] == 14) | (sf["hour"] == 15) | (sf["hour"] == 16)] = "Day (11-16)"

display(sf["Timeperiod"].value_counts())
```

I would've expected more crime during the night, but on the other hand, most people are asleep at night and usually cant report crimes there or be a victim. To confirm that, we can group category plots by the period:

```python
sns.catplot(x="Category", col="Timeperiod", data=sf, kind="count", col_wrap=1,
           aspect=1.3)
plt.xticks(rotation="vertical")
plt.show()
```

This is really messy. A lot of crimes don't occur very often (look at the list above), so I will drop them from the dataframe, to get a clearer picture of the more prevalent crimes. Everything below 1000 records will be dropped, except robbery:

```python
droplist = sf.loc[(sf["Category"] == "ARSON") |
                 (sf["Category"] == "RUNAWAY") |
                 (sf["Category"] == "LIQUOR LAWS") |
                 (sf["Category"] == "DISORDERLY CONDUCT") |
                 (sf["Category"] == "FORGERY/COUNTERFEITING") |
                 (sf["Category"] == "VANDALISM") |
                 (sf["Category"] == "SUICIDE") |
                 (sf["Category"] == "FAMILY OFFENSES") |
                 (sf["Category"] == "EMBEZZLEMENT") |
                 (sf["Category"] == "STOLEN PROPERTY") |
                 (sf["Category"] == "EXTORTION") |
                 (sf["Category"] == "BURGLARY") |
                 (sf["Category"] == "LOITERING") |
                 (sf["Category"] == "PORNOGRAPHY/OBSCENE MAT") |
                 (sf["Category"] == "GAMBLING") |
                 (sf["Category"] == "DRIVING UNDER THE INFLUENCE") |
                 (sf["Category"] == "PROSTITUTION") |
                 (sf["Category"] == "KIDNAPPING") |
                 (sf["Category"] == "FRAUD") |
                 (sf["Category"] == "DRUNKENNESS") |
                 (sf["Category"] == "TRESPASS") |
                 (sf["Category"] == "WEAPON LAWS") |
                 (sf["Category"] == "SECONDARY CODES") |
                  (sf["Category"] == "ROBBERY") |
                 (sf["Category"] == "BRIBERY")].index

alt_sf = sf.drop(droplist)
alt_sf["Category"].value_counts()
```

```python
cat = sns.catplot(x="Category", col="Timeperiod", 
                  data=alt_sf, 
                  kind="count", 
                  col_wrap=2, 
                  aspect=1,
                 col_order = ["Morning (5-10)", "Day (11-16)", "Evening (17-22)", "Night (23-4)"])
cat.set_xticklabels(rotation="vertical")
plt.show()
```

It is still a little bit messy, but you can see a lot more in them, than before. For example that there is a high fluctuation in Larceny & Theft throughout the day. But considering that there are more people outside in the evening and during the day to steal from, this is not unusual.

The same pattern can be seen for the other crime categories.
